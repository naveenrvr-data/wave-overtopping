# Scaling Input Features

```python
#Froude scaling law
# Scaling features with unit 'm' using significant wave height feature Hm0 toe
W1[['Hm0 d','hdeep','h','ht','Bt','D', 'Rc', 'B', 'hb','Bh','Ac','Gc']] = W1[['Hm0 d','hdeep','h','ht',
                                                                               'Bt','D', 'Rc', 'B', 'hb','Bh','Ac','Gc']].div(W1['Hm0 toe'],axis = 0)
```

```python
# Dividing wave period using sqrt(hm0 toe) to make it dimensionless- wave energy pdf page 7
W1[['Tp d', 'Tm d', 'Tm1,0 d','Tp toe', 'Tm toe', 'Tm1,0t']] =  W1[['Tp d', 'Tm d', 'Tm1,0 d',
                                                                    'Tp toe', 'Tm toe', 'Tm1,0t']].divide(np.sqrt(W1['Hm0 toe']), axis=0)
```
